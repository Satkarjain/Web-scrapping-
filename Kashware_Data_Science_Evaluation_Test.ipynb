```
from requests import get
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import pandas as pd
import re
import re
import time
```

```
def dict_maker():
  offer_details = {}
  offer_details['Offer_heading'] = []
  offer_details['By '] = []
  offer_details['Bank  '] = []
  offer_details['Valid Till '] = []
  offer_details['Deal Type'] = []
  offer_details['Card Type'] = []
  offer_details['Segment'] = []
  offer_details['Info'] = []
  return offer_details
```

```
options = webdriver.ChromeOptions()
options.add_argument("-headless")
options.add_argument("-no-sandbox")
options.add_argument("-disable-dev-shm-usage")
```

```
offer_details = dict_maker()
df= pd.DataFrame()
for x in range(1,4):
  url = "https://happycredit.in/kotak-mahindra-bank-offers/?page="+str(x)
  browser =webdriver.Chrome("chromedriver",options=options)
  browser.get(url)
  time.sleep(3)
  html = browser.page_source
  html_soup = BeautifulSoup(html, 'lxml')
  offer_list = html_soup.find_all('a', class_ = "coupon-cart others")
  for offer in offer_list:
    offer_details["Offer_heading"].append(offer.h3.text)
    offer_details["Info"].append(offer.find('p',class_ = 'info-txt').text)
    other_details = offer.find_all('span')
    flag = 0
    for f in other_details:
      a= f.text.split(':')
      if a[0] == "Segment":
        flag =1
      offer_details[a[0]].append(a[1])
    
    if flag == 0:
      offer_details['Segment'].append('Not given')
```

```
df = pd.DataFrame(offer_details)
```

```
def new_offer_addition(df,x,url):       # "x" is the number of new offers added in that page
  offer_details = dict_maker()                    
  browser =webdriver.Chrome("chromedriver",options=options)
  browser.get(url)
  time.sleep(3)
  html = browser.page_source
  html_soup = BeautifulSoup(html, 'lxml')
  offer_list = html_soup.find_all('a', class_ = "coupon-cart others")
  len1 = len(offer_list)
  while x>0:
    offer = offer_list[len1-x]
    x= x-1
    offer_details["Info"].append(offer.find('p',class_ = 'info-txt').text)
    offer_details["Offer_heading"].append(offer.h3.text)
    other_details = offer.find_all('span')
    
    for f in other_details:
      flag = 0
      a= f.text.split(':')
      if a[0] == "Segment":
        flag =1
      offer_details[a[0]].append(a[1])

    if flag == 0:
      offer_details['Segment'].append('Not given')
  
  df1 = pd.DataFrame(offer_details)
  df = pd.concat([df,df1])
  return df
```

```
df = new_offer_addition(df,2,"https://happycredit.in/kotak-mahindra-bank-offers/?page=3")        #It will add new "x" items to df
```
